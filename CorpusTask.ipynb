###1.2 Очистка от html тегов и загрузка текстов

````python
import requests
from bs4 import BeautifulSoup
import re
import time
from tqdm import tqdm
# Specify the URL of the website you want to scrape
url = 'http://www.lib.ru/BUNIN/'

# Send a GET request to the website
response = requests.get(url)

# Parse the HTML content of the page with BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find all links on the page
links = soup.find_all('a', href=re.compile("\.txt$"))

texts = []

for link in tqdm(filtered_links):
    time.sleep(1)
    response = requests.get(url + link['href'])  # Make sure to use the full URL if needed
    soup = BeautifulSoup(response.content, 'html.parser')

    # Remove unwanted tags
    for tag in ['script', 'style', 'ul']:
        for element in soup(tag):
            element.decompose()  # Removes the tag

    # Preserve text style tags
    for tag in ['i', 'b', 'u', 's']:
        for element in soup(tag):
            element.unwrap()  # Unwraps the tag, leaving only its content
    # Remove <div align=right> and its contents
    div_right = soup.find('div', {'align': 'right'})
    if div_right:
        div_right.decompose()

    # Remove <form> and its contents
    form = soup.find('form')
    if form:
        form.decompose()

    # Remove <select> and its contents
    select = soup.find('select')
    if select:
        select.decompose()

    # Remove <option> tags
    options = soup.find_all('option')
    for option in options:
        option.decompose()

    # Remove <input type="submit">
    input_submit = soup.find('input', {'type': 'submit'})
    if input_submit:
        input_submit.decompose()

    # Remove <small> tag
    small = soup.find('small')
    if small:
        small.decompose()
    # Remove <h2> tags
    h2_tags = soup.find_all('head')
    for h2 in h2_tags:
        h2.decompose()

    text = soup.get_text(separator=' ')
    texts.append(text)
````

Были убраны ненужные html элементы (рейтинг текста, дата последней модификации) и заглавия

````python
for text in texts:
    print(text[:500])
    print("." * 50)
    print(text[-500:])
    print('*' * 50)
````

###1.3 Очистка текстов

````python
import re

def contains_digit(s):
    return any(c.isdigit() for c in s)

def clean_text(text):
    lines = text.split('\n')
    cleaned_text = ''
    for line in lines:
        if len(line) == 0:
            continue

        if line[-1] in ['-', ' ', ')']:
            continue
        if "OCR" in line:
            continue
        if "файл" in line:
            continue
        if contains_digit(line):
            continue
        if "Бунин" in line:
            continue
        if "Компьютерный" in line:
            continue
        if "Оригинал этого текста расположен в библиотеке" in line:
            continue
        cleaned_text += line + '\n'

    return cleaned_text
````

````python
cleaned_texts = [clean_text(text) for text in texts]
````

````python
def process_strings(cleaned_texts):
    # Replace all '\n' with spaces in each string
    processed_texts = [text.replace('\n', ' ') for text in cleaned_texts]
    # Join all processed texts into one string, removing extra spaces
    merged_text = ' '.join(processed_texts)
    merged_text = ' '.join(merged_text.split())

    return merged_text
````
